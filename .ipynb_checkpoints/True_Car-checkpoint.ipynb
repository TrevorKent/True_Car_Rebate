```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "data/DataSet2.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd.head(10)
```

```python
x_axis = data_file_pd["Month"]
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(50,15))
plt.bar(x_axis, data_file_pd["Incentive"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, data_file_pd["Month"], rotation="vertical")
plt.show()
```
