```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "data/DataSet2.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd.head(10)
```

```python
x_axis = data_file_pd["Month"]
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(15, 6))
plt.bar(x_axis, data_file_pd["Incentive"], color='r', alpha=0.5, align="center", label="Incentive")
plt.xticks(tick_locations, data_file_pd["Month"], rotation="50", fontsize="xx-small")
plt.title("Average Vehicle Incentive Spend ($)", fontsize="xx-large")
plt.ylim(2000, 3200)
plt.xlabel("Months", fontsize="x-large")
plt.ylabel("Average Vehicle Incentive Spend ($)", fontsize="x-large")
plt.legend(loc="lower right")
plt.margins(x=.005)
plt.show()
```

```python
plt.savefig("Image")
```

```python
# Using the class method gives us even more control
fig, ax = plt.subplots(figsize=(20, 10))

ax2 = ax.twinx()
ax.bar(x_axis, data_file_pd["Incentive"], alpha=.5)
ax2.plot(x_axis, data_file_pd["% TP"]*100, color='red', linewidth=4)

plt.setp(ax.get_xticklabels(), rotation=90, ha="center")

ax.set_ylim(2000, 3200)
ax2.set_ylim(4, 12)

ax.set_ylabel("Average Vehicle Incentive Spend ($)", fontsize="xx-large")
ax2.set_ylabel("% Transaction Price", fontsize="xx-large")

plt.margins(x=.005)
plt.title("Average Montly Incentive Spend ($)", fontsize="xx-large")
ax.set_xlabel('Month', fontsize="xx-large")

plt.legend(fontsize="x-large")
plt.show()
```
