```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "data/DataSet2.csv"
```

```python
data_file_pd = pd.read_csv(data_file)
data_file_pd.head(10)
```

```python
x_axis = data_file_pd["Month"]
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(14,8))
plt.bar(x_axis, data_file_pd["Incentive"], color='r', alpha=0.5, align="center", label="Incentive")
plt.xticks(tick_locations, data_file_pd["Month"], rotation="vertical", fontsize="xx-small")
plt.title("Average Montly Incentive Spend", fontsize="xx-large")
plt.ylim(2000, 3200)
plt.xlabel("Months", fontsize="x-large")
plt.ylabel("Dollars", fontsize="x-large")
plt.legend(loc="lower right")
plt.margins(x=.005)
plt.show()
```

```python
plt.savefig("Image")
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x_axis, data_file_pd["Incentive"])
ax2.plot(x_axis, data_file_pd["% TP"], color='red')
ax1.set_ylim(2000, 3200)
ax2.set_ylim(.04, .12)
plt.margins(x=.005)
plt.show()
```
